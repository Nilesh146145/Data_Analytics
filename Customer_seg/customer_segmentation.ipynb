Outliers

```python
Q1=data['MntRegularProds'].quantile(0.25)
Q3=data['MntRegularProds'].quantile(0.75)
IQR=Q3-Q1
Lower_fence=Q1-1.5*IQR
Upper_fence=Q3+1.5*IQR
Outliers=data[(data['MntRegularProds']<Lower_fence) | (data['MntRegularProds']>Upper_fence)]
Outliers.head()
```

Outliers removal

```python
data=data[(data['MntRegularProds']>Lower_fence) & (data['MntRegularProds']<Upper_fence)]
data.describe()
```

```python
def get_marital_status(row):
    if row['marital_Divorced']==1:
        return 'Divorced'
    elif row['marital_Married']==1:
        return 'Married'
    elif row['marital_Single']==1:
        return 'Single'
    elif row['marital_Together']==1:
        return 'Together'
    elif row['marital_Widow']==1:
        return 'Widow'
    else:
        return 'Unknown'
data['Marital']=data.apply(get_marital_status, axis=1)
```

K-Means Clustering
-Standardising data
-Principal Component Analysis (PCA)
-Elbow method
-Silhouette score analysis

Standardising data

```python
from sklearn.cluster import KMeans
```

```python
from sklearn.preprocessing import StandardScaler , LabelEncoder
scaler = StandardScaler()
le = LabelEncoder()
data['Marital_Encoded']=le.fit_transform(data['Marital'])
column_for_clustering=['Income','Age','MntRegularProds','Marital_Encoded']
scaled_data=data.copy()
scaled_data[column_for_clustering]=scaler.fit_transform(data[column_for_clustering])
scaled_data[column_for_clustering].describe()
```

Principal Component Analysis(PCA)

```python
from sklearn import decomposition
pca = decomposition.PCA(n_components=2)
pca_res= pca.fit_transform(scaled_data[column_for_clustering])
scaled_data['pc1']= pca_res[:,0]
scaled_data['pc2']=pca_res[:,1]
```

Elbow method

```python
X = scaled_data[column_for_clustering]
inertia_list = []
for K in range(2,10):
    inertia=KMeans(n_clusters=K, random_state=7).fit(X).inertia_
    inertia_list.append(inertia)
```

```python
plt.figure(figsize=[7,10])
plt.plot(range(2,10), inertia_list, color=(54/255, 113/255, 130/255))
plt.title("Inertia vs. Number of Clusters")
plt.xlabel("Number of Clusters(K)")
plt.ylabel("Inertia")
plt.show()
```

Silhouette score analysis

```python
from sklearn.metrics import silhouette_score
silhouette_list = []
for K in range(2,10):
    model = KMeans(n_clusters=K, random_state=7)
    clusters = model.fit_predict(X)
    s_avg = silhouette_score(X, clusters)
    silhouette_list.append(s_avg)

plt.figure(figsize=[7,10])
plt.plot(range(2,10), silhouette_list, color=(54/255, 113/255, 130/255))
plt.title(" Silhouette Score vs Number of Clusters")
plt.xlabel("Number of Clusters(K)")
plt.ylabel(" Silhouette Score")
plt.show()
```

```python
model = KMeans(n_clusters=5, random_state=7)
model.fit(scaled_data[column_for_clustering])
scaled_data['Cluster'] = model.predict(scaled_data[column_for_clustering])
```
